## 1.5.1 Bluecar_counter

```
# Read our cleaned csv file

#cars = pd.read_csv('cars.csv')
#cars.head()
```

```
# create a new column where we find the difference in the number of cars at the station

al['bluecar_diff'] = al['bluecar_counter'].diff()
al.head()
```

```
# Challenge 1
# What is the most popular hour of the day for picking up a shared electric car

# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

# positive returned

# Negative taken

#select city == Paris

al_paris = al[al['city'] =='Paris']

al_paris
```

```
al_paris[al_paris['bluecar_diff'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```
# Challenge 2
# What is the most popular hour for returning cars

al_paris[al_paris['bluecar_diff'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```
# Challenge 3
# What is the most popular station overall

al_paris[(al_paris['kind'] == 'STATION') & (al_paris['status'] == 'ok')].groupby('public_name')['public_name'].count().sort_values(ascending= False)
```

```
# Challenge 4
# What is the most popular station at most popular picking hours

al_paris[(al_paris['kind'] == 'STATION') & (al_paris['status'] == 'ok') & (al_paris['hour'] == 4)].groupby('public_name')['public_name'].count().sort_values(ascending= False)
```

```
# Challenge 5
# What is the most popular postal code for picking up blue cars 

#al_paris.groupby('postal_code').count()[['bluecar_counter']].head(5)

al_paris[(al_paris['postal_code']) & (al_paris['status'] == 'ok')].groupby('postal_code')['postal_code'].count().sort_values(ascending= False)
```

```
# Does the station in challenge 4 belong to our postal code??Overall

al_paris[(al_paris['kind'] == 'STATION') & (al_paris['status'] == 'ok') & (al_paris['hour'] == 4)].groupby('postal_code')['postal_code'].count().sort_values(ascending= False)
```

```
# Does the station in challenge 4 belong to our postal code??Most popular picking hours

al_paris[(al_paris['kind'] == 'STATION') & (al_paris['status'] == 'ok') & (al_paris['hour'] == 4)].groupby('postal_code')['postal_code'].count().sort_values(ascending= False)
```
